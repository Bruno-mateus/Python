### Hands-On - DistribuicÃßaÃÉo em variaÃÅveis numeÃÅricas


```python
## histograma do tempo de contrato 
df_churn.tenure.plot.hist()
```

- A vari√°vel tempo de contrato (tenure) n√£o apresenta visualmente uma distribui√ß√£o normal.
- A grande  dos valores se concentram nos extremos, ou seja, contratos com poucos meses de ativa√ß√£o e contratos logns, acima aproxidamente dos 65 meses.

```python
## histograma do monthly charges
## representa o valor pago mensalmente dos contratos dos clientes
df_churn.MonthlyCharges.plot.hist()
## √© possivel analisar que a maioria dos contratos feitos s√£o do menor valor
```

```python
## medidas de posi√ß√£o - Mediana do tempo de contrato
## mediana n√£o esta suejeita a outlier, diferente da m√©dia ela n√£o √© sensivel a valores extremos
df_churn.tenure.median()
```

```python
# medidas de posi√ß√£o - Moda do tempo de contrato
# o tempo de contrato mais frequente na mminha base
df_churn.tenure.mode()
```

```python
# medidas de disper√ß√£o - desvio padr√£o do tempo de contrato
# os valores que variam pra cima em rela√ß√£o a m√©dia media durante os meses
df_churn.tenure.std()
```

```python
# medidas de dispers√£o - coeficiente de varia√ß√£o
# A varia√ß√£o coeficiente o tempo de contrato
df_churn.tenure.std()/df_churn.tenure.mean() * 100
```

## filtros e agrupamentos de valores

```python
## filtrando todos os clientes com contem apenas 1 mes de contrarto
df_churn[(df_churn.tenure==1)]
```

```python
## total e percentual desses clientes com um m√™s dde contrato
cTenureOneMon = len(df_churn[(df_churn.tenure==1)])
cTenureAllTime = len(df_churn['tenure'])
print("total de clientes:")
print(cTenureAllTime)
print("Total de clientes com um m√™s de contrato:")
print(cTenureOneMon)
print("% total de clientes com um m√™s de contrato")
print((cTenureOneMon/cTenureAllTime)*100)
```

```python
    ## apresentar a quantidade de clientes por tempo de contrato - Agrupamento com sumariza√ß√£o
df_churn.groupby(['tenure'])['tenure'].count().sort_values(ascending=False)
```

```python
# apresentar a quantidade de clientes por tempo de contrato - Agrupamento com sumariza√ß√£o em grafico area

tenurePlot = (df_churn.groupby(["tenure"])["tenure"].count()).plot.area()
```

```python
# apresentar a quantidade de clientes por tempo de contrato - Agrupamento com sumariza√ß√£o em grafico barh
                                                                    ## aumenta o gr√°fico em polegadas
tenurePlot2 = (df_churn.groupby(["tenure"])["tenure"].count().sort_values()).plot.barh(figsize=(21,21))
tenurePlot2.bar_label(tenurePlot2.containers[0])
```

```python
# Clientes com contrato do tipo mensal s√£o mais propensos ao churn
# construir tabela de contigencia entre tipo de contratos e churn
pd.crosstab(df_churn.Churn,df_churn.Contract,margins=True,normalize='index', margins_name="Total")
```

Considerando o publico total, os clientes de contrato mensal repesantao 55%, por√©m se considerarmos apenas os clientes que abondonaram o servi√ßo 88% tinham contraro mensal. Desta forma, podemos afirmar  que h√° uma corre√ß√£o entre o tipo de contrato mensal e o abandono
