# Lidando com valores ausentes

```python
## Remover de forma especifica e direta uma coluna q a qual ja sabemos que h√° valores ausentes
df_churn.drop(columns=['TotalCharges'])
```

```python
## Remover qualquer linha ou coluna que contem algum valor vazio qualquer
# 0 = linnha
# 1 = coluna
df_churn.dropna(axis=1)
```

```python
## Remover qualquer linha ou coluna que contem TODOS os velores vazios
# 0 = linnha
# 1 = coluna
df_churn.dropna(axis=1, how='all')
```
