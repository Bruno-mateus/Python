### Hands-On - Distribuição em variáveis numéricas


```python
## histograma do tempo de contrato 
df_churn.tenure.plot.hist()
```

- A variável tempo de contrato (tenure) não apresenta visualmente uma distribuição normal.
- A grande  dos valores se concentram nos extremos, ou seja, contratos com poucos meses de ativação e contratos logns, acima aproxidamente dos 65 meses.

```python
## medidas de posição - Mediana do tempo de contrato
## mediana não esta suejeita a outlier, diferente da média ela não é sensivel a valores extremos
df_churn.tenure.median()
```

```python
# medidas de posição - Moda do tempo de contrato
# o tempo de contrato mais frequente na mminha base
df_churn.tenure.mode()
```

```python
# medidas de disperção - desvio padrão do tempo de contrato
# os valores que variam pra cima em relação a média media durante os meses
df_churn.tenure.std()
```

```python
# medidas de dispersão - coeficiente de variação
# A variação coeficiente o tempo de contrato
df_churn.tenure.std()/df_churn.tenure.mean() * 100
```

## filtros e agrupamentos de valores

```python
## filtrando todos os clientes com contem apenas 1 mes de contrarto
df_churn[(df_churn.tenure==1)]
```

```python
## total e percentual desses clientes com um mês dde contrato
cTenureOneMon = len(df_churn[(df_churn.tenure==1)])
cTenureAllTime = len(df_churn['tenure'])
print("total de clientes:")
print(cTenureAllTime)
print("Total de clientes com um mês de contrato:")
print(cTenureOneMon)
print("% total de clientes com um mês de contrato")
print((cTenureOneMon/cTenureAllTime)*100)
```

```python
    ## apresentar a quantidade de clientes por tempo de contrato - Agrupamento com sumarização
df_churn.groupby(['tenure'])['tenure'].count().sort_values(ascending=False)
```

```python
# apresentar a quantidade de clientes por tempo de contrato - Agrupamento com sumarização em grafico area

tenurePlot = (df_churn.groupby(["tenure"])["tenure"].count()).plot.area()
```

```python
# apresentar a quantidade de clientes por tempo de contrato - Agrupamento com sumarização em grafico barh
                                                                    ## aumenta o gráfico em polegadas
tenurePlot2 = (df_churn.groupby(["tenure"])["tenure"].count().sort_values()).plot.barh(figsize=(21,21))
tenurePlot2.bar_label(tenurePlot2.containers[0])
```

```python
# Clientes com contrato do tipo mensal são mais propensos ao churn
# construir tabela de contigencia entre tipo de contratos e churn
pd.crosstab(df_churn.Churn,df_churn.Contract,margins=True,normalize='index', margins_name="Total")
```

Considerando o publico total, os clientes de contrato mensal repesantao 55%, porém se considerarmos apenas os clientes que abondonaram o serviço 88% tinham contraro mensal. Desta forma, podemos afirmar  que há uma correção entre o tipo de contrato mensal e o abandono
